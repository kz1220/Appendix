```python
import matplotlib.pyplot as plt
import numpy as np

points = [(0, 0), (1, 1), (3, 1)]
# Plot the shaded area
plt.fill_between([0,1,3], [0,1,1], color="grey", alpha=0.5)


#plot solid line
plt.plot([1,3],[1,1],color='black' )

#plot dashed line
plt.plot([0,1],[0,1],linestyle='dashed',color='black')

# Plot the solid dot at c
plt.plot(1, 1, 'ko')

# Plot the open circle at B1(p2)
plt.plot(2, 2, 'ko', fillstyle='none', markeredgewidth=2)

#vertical line at p^m
plt.vlines(2, 2.04, 3, colors='black')

plt.text(2, 1.5, '$BR_1(p_2)$')
plt.xlabel('$p_1$')
plt.ylabel('$p_2$')
plt.xlim(0, 3)
plt.ylim(0, 3)

# Set the x-axis ticks
plt.xticks([0, 1, 2], ['0', '$c$', '$p^m$'])

# Set the y-axis ticks
plt.yticks([1,2], ['$c$','$p^m$'])

# Show the plot
plt.show()
```
